```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
# Use Map to format all the columns
# formats a number and converts to currency rounding to 2 dec points.
#.map("${:.2f}".format)
# formats a number and converts to number with , separations
#.map("{:,}".format)
# formats a number and rounding to 2 points.
#.map("{:.2f}".format)


total_school = school_data_complete["school_name"].nunique()
total_students = school_data_complete["student_name"].nunique()
total_budget = "${:,.2f}".format(school_data_complete["budget"].sum())
average_ms = school_data_complete["math_score"].mean()
average_rs = school_data_complete["reading_score"].mean()
average_pr = (average_ms + average_rs) / 2
perc_pass_ms = len(school_data_complete.loc[school_data_complete["math_score"] >= 70])/total_students
perc_pass_rs = len(school_data_complete.loc[school_data_complete["reading_score"] >= 70])/total_students

# Make the dataframe
dist_df = pd.DataFrame([{'Total Schools': total_school,
                        'Total Students': total_students,
                        'Total Budget': total_budget,
                        'Average Math Score': average_ms,                       
                        'Average Reading Score': average_rs,  
                        'Overall Pass Rate': average_pr,  
                        '% of Students Math Score Above 70': perc_pass_ms,
                        '% of Students Reading Score Above 70': perc_pass_rs
                        }])


All_schools_df = dist_df[["Total Schools","Total Students","Total Budget","Average Math Score", "Average Reading Score", "Overall Pass Rate", "% of Students Math Score Above 70", "% of Students Reading Score Above 70"]]
All_schools_df.head()
```

```python
pass_ms = school_data_complete.loc[school_data_complete["math_score"] >= 70]
pass_rs = school_data_complete.loc[school_data_complete["reading_score"] >= 70]

by_school_pms = pass_ms.groupby('school_name').agg({'math_score':'count'})
by_school_prs = pass_rs.groupby('school_name').agg({'reading_score':'count'})


pass_score_merged = pd.merge(by_school_pms, by_school_prs, on='school_name', how='outer')

summarized = pd.merge(by_school_df, pass_score_merged, on='school_name', how='outer')
summarized['% Passing Math'] = summarized['math_score_y']/summarized['student_name'] 
summarized['% Passing Reading'] = summarized['reading_score_y']/summarized['student_name']
summarized['% Passing'] = (summarized['% Passing Math']  + summarized['% Passing Reading'])  / 2

summarized = summarized.rename(columns={"student_name": "Total_Students", 
                                        "budget": "Total_Budget",
                                        "math_score_x": "Average Math Score",
                                        "reading_score_x": "Average Reading Score", 
                                        "student budget": "Budget per Student", 
                                        "math_score_y": "Passing Math",
                                        "reading_score_y": "Passing Reading"})

summarized
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school

```python
nineth = school_data_complete[(school_data_complete["grade"]=="9th")]
tenth = school_data_complete[(school_data_complete["grade"]=="10th")]
elevth = school_data_complete[(school_data_complete["grade"]=="11th")]
twelth = school_data_complete[(school_data_complete["grade"]=="12th")]

# group by 
by_school_9th = nineth.groupby('school_name').agg(
    {'math_score':'mean'})
by_school_10th = tenth.groupby('school_name').agg(
    {'math_score':'mean'})
by_school_11th = elevth.groupby('school_name').agg(
    {'math_score':'mean'})
by_school_12th = twelth.groupby('school_name').agg(
    {'math_score':'mean'})

## Combine the series by using merge - use suffix so you dont have to rename

m9and10 = pd.merge(by_school_9th, by_school_10th, on='school_name', how='outer', suffixes =('_9th', '_10th'))
m11and12 = pd.merge(by_school_11th, by_school_12th, on='school_name', how='outer', suffixes =('_11th', '_12th'))

m9and12 = pd.merge(m9and10, m11and12, on='school_name', how='outer')
m9and12
```
